```python
import seaborn as sns
import matplotlib.pyplot as plt
import os
import pandas as pd
from scipy.stats import mannwhitneyu
import statsmodels.api as sm
import statsmodels.formula.api as smf
from scipy.stats import spearmanr, pearsonr
import matplotlib as mpl

mpl.rcParams["pdf.fonttype"] = 42   # TrueType fonts (editable)
mpl.rcParams["ps.fonttype"] = 42
mpl.rcParams["svg.fonttype"] = "none"
# sns.set(style="whitegrid")

# ----------------------------------------
# 1. Create outcome group
# ----------------------------------------
nfl_plot_data['mrs_90_bin'] = nfl_plot_data['mrs_90'].apply(
    lambda x: "Good (0–3)" if x <= 3 else "Poor (4–6)"
)

# consistent palette
palette = {
    "Good (0–3)": sns.color_palette("tab10")[0],
    "Poor (4–6)": sns.color_palette("tab10")[1],
}

# ----------------------------------------
# 2. Jitter + outcome-based x separation
# ----------------------------------------
jitter_width = 0.10   # small jitter for readability
group_offset = 0.15   # separation between groups

rng = np.random.default_rng(42)
nfl_plot_data_current = nfl_plot_data.copy()
nfl_plot_data_current = nfl_plot_data_current.dropna(subset=['mrs_90'])
# jitter
nfl_plot_data_current['x_jitter'] = rng.uniform(
    -jitter_width, jitter_width, size=len(nfl_plot_data_current)
)

# group offset
group_map = {
    "Good (0–3)": -group_offset,
    "Poor (4–6)": +group_offset
}

nfl_plot_data_current['x_plot'] = nfl_plot_data_current.apply(
    lambda row: row['days_after_lkn'] +
                row['x_jitter'] +
                group_map[row['mrs_90_bin']],
    axis=1
)

# ----------------------------------------
# 3. Summary Counts (subjects per day per outcome)
# ----------------------------------------
summary_counts = (
    nfl_plot_data_current.groupby(['days_after_lkn', 'mrs_90_bin'])['personid']
    .nunique()
    .reset_index()
    .pivot(index='days_after_lkn', columns='mrs_90_bin', values='personid')
)

print("\nNumber of subjects per day by outcome group:")
print(summary_counts.fillna(0))

# ----------------------------------------
# 4. Compute group mean ± SD (use true day positions)
# ----------------------------------------
summary_stats = nfl_plot_data_current.groupby(['days_after_lkn', 'mrs_90_bin']).agg(
    mean=('log_nfl', 'mean'),
    sd=('log_nfl',  'std')
).reset_index()

# add x-position offset for group separation (WITHOUT jitter)
summary_stats['x_mean'] = summary_stats.apply(
    lambda row: row['days_after_lkn'] + group_map[row['mrs_90_bin']],
    axis=1
)

# ----------------------------------------
# 5. Plot
# ----------------------------------------
plt.figure(figsize=(11, 6))

# A. Jittered raw points
sns.scatterplot(
    data=nfl_plot_data_current,
    x='x_plot', y='log_nfl',
    hue='mrs_90_bin',
    palette=palette,
    s=55,
    alpha=0.85,
    linewidth=0.5,
    edgecolor='black'
)

# B. Individual subject lines (perfect alignment)
for (pid, group), dfsub in nfl_plot_data_current.groupby(['personid', 'mrs_90_bin']):
    plt.plot(
        dfsub['x_plot'],
        dfsub['log_nfl'],
        color=palette[group],
        alpha=0.25,
        linewidth=1
    )

# C. Mean ± SD (plotted at true timepoints + group offset)
for group, dfsub in summary_stats.groupby('mrs_90_bin'):
    col = palette[group]

    # mean line
    plt.plot(
        dfsub['x_mean'], dfsub['mean'],
        color=col,
        linewidth=3
    )

    # SD ribbon
    plt.fill_between(
        dfsub['x_mean'],
        dfsub['mean'] - dfsub['sd'],
        dfsub['mean'] + dfsub['sd'],
        color=col,
        alpha=0.18
    )

# ----------------------------------------
# 6. Formatting
# ----------------------------------------
ax = plt.gca()

# Thicken visible spines
for spine in ["bottom", "left"]:
    ax.spines[spine].set_linewidth(1.8)

# Optionally hide top/right spines (clean, journal style)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.tick_params(
    axis="both",
    which="major",
    length=7,      # tick length
    width=1.6,     # tick thickness
    labelsize=11
)
plt.title("Longitudinal NfL Values by Functional Outcome (mRS-90)", fontsize=16)
plt.xlabel("Days After Stroke", fontsize=12)
plt.ylabel("log(NfL)", fontsize=12)

# Legend formatting
plt.legend(title="Outcome", frameon=True)

plt.tight_layout()
fig_dir = '/Users/ganshengtan/Library/CloudStorage/Box-Box/Washu/projects/taVNS_cytokine/stroke-taVNS/nfl_study/figs/'


pdf_path = os.path.join(fig_dir, f"mrs_90_nfl.pdf")
eps_path = os.path.join(fig_dir, f"mrs_90_nfl.eps")
plt.savefig(pdf_path, format="pdf")
plt.savefig(eps_path, format="eps")
plt.show()



print("gee, dependent 1 is good")
gee = smf.gee(
    "mrs_90_bin ~ log_nfl",
    groups=nfl_plot_data_current["personid"],
    data=nfl_plot_data_current,
    family=sm.families.Binomial()
).fit()

print(gee.summary())

# nfl_plot_data["mrs_90_bin_num"] = nfl_plot_data["mrs_90"].apply(lambda x: 0 if x <= 3 else 1)
nfl_plot_data["mrs_90_bin_num"] = np.where(
    nfl_plot_data["mrs_90"].isna(),
    np.nan,
    (nfl_plot_data["mrs_90"] > 3).astype(int)
)
logit = smf.logit("mrs_90_bin_num ~ log_nfl", nfl_plot_data.dropna(subset=["log_nfl", "mrs_90"])).fit()
print(logit.summary())

# daywise_results = pd.DataFrame(daywise_results,
#     columns=["day", "n_good", "n_poor", "p_value"])

# print(daywise_results)


clean = nfl_plot_data[['log_nfl','mrs_90', 'mrs_90_bin', 'mrs_90_bin_num', 'days_after_lkn']].dropna(subset=["log_nfl", "mrs_90"])

rho, p_rho = spearmanr(clean['log_nfl'], clean['mrs_90_bin_num'])
r, p_r  = pearsonr(clean['log_nfl'], clean['mrs_90_bin_num'])

print("Spearman:", rho, p_rho)

rho_all, p_s_all = spearmanr(clean["log_nfl"], clean["mrs_90"])
r_all, p_p_all = pearsonr(clean["log_nfl"], clean["mrs_90"])

print("\n=== Overall Correlations ===")
print(f"Spearman rho = {rho_all:.3f}, p = {p_s_all:.4f}")
print(f"Pearson r = {r_all:.3f}, p = {p_p_all:.4f}")

daywise_corrs = []

for day, df_day in clean.groupby("days_after_lkn"):
    if df_day["mrs_90"].nunique() > 1:
        rho, p_s = spearmanr(df_day["log_nfl"], df_day["mrs_90"])
        r, p_p = pearsonr(df_day["log_nfl"], df_day["mrs_90"])
    else:  # cannot compute correlation if all mRS same
        rho, p_s, r, p_p = np.nan, np.nan, np.nan, np.nan

    daywise_corrs.append([day, len(df_day), rho, p_s, r, p_p])

daywise_corrs = pd.DataFrame(
    daywise_corrs,
    columns=["day", "n", "Spearman_rho", "Spearman_p", "Pearson_r", "Pearson_p"]
)

print("\n=== Day-wise Correlations (log_nfl vs mRS 90) ===")
print(daywise_corrs)

from scipy.stats import mannwhitneyu, ttest_ind, pointbiserialr
import numpy as np

results = []

for day, df_day in nfl_plot_data.groupby("days_after_lkn"):

    df_day = df_day.dropna(subset=['log_nfl', 'mrs_90_bin_num'])
    g = df_day[df_day['mrs_90_bin_num'] == 0]['log_nfl']
    p = df_day[df_day['mrs_90_bin_num'] == 1]['log_nfl']

    # Skip day if any group has <2 subjects
    if len(g) < 2 or len(p) < 2:
        results.append([day, len(g), len(p),
                        np.nan, np.nan, np.nan, np.nan, np.nan, np.nan])
        continue

    # ------------------------------------------------------
    # A. Mann–Whitney U
    # ------------------------------------------------------
    U, p_mwu = mannwhitneyu(g, p, alternative="two-sided")

    # Rank-biserial correlation
    n1, n2 = len(g), len(p)
    r_rb = 1 - (2 * U) / (n1 * n2)

    # ------------------------------------------------------
    # B. t-test
    # ------------------------------------------------------
    t_stat, p_t = ttest_ind(g, p, equal_var=False)

    # Cohen’s d (Hedges' g optional)
    mean_diff = g.mean() - p.mean()
    pooled_sd = np.sqrt(((n1 - 1)*g.var() + (n2 - 1)*p.var()) / (n1 + n2 - 2))
    d = mean_diff / pooled_sd

    # ------------------------------------------------------
    # C. Point-biserial correlation
    # ------------------------------------------------------
    labels = df_day['mrs_90_bin_num']     # 0 good, 1 poor
    values = df_day['log_nfl']
    r_pb, p_pb = pointbiserialr(labels, values)

    # ------------------------------------------------------
    # Store
    # ------------------------------------------------------
    results.append([
        day, n1, n2,
        U, p_mwu,
        t_stat, p_t,
        d, r_rb, r_pb
    ])

# Final dataframe
results_df = pd.DataFrame(
    results,
    columns=[
        'day','n_good','n_poor',
        'U','p_mwu',
        't_stat','p_t',
        'cohen_d','rank_biserial_r','point_biserial_r'
    ]
)

print("\n=== Day-wise Effect Size Summary ===")
print(results_df)
```

```python
import os
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

from scipy.stats import spearmanr, kendalltau
from numpy.linalg import lstsq


fig_dir = (
    "/Users/ganshengtan/Library/CloudStorage/Box-Box/"
    "Washu/projects/taVNS_cytokine/stroke-taVNS/nfl_study/figs/"
)


def finalize_and_save(fig_name, fig=None):
    """
    Apply final axis formatting (for current axes) and save.
    If fig is provided, saves that fig; else saves current plt figure.
    """
    ax = plt.gca()

    for spine in ["bottom", "left"]:
        ax.spines[spine].set_linewidth(1.8)
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)

    ax.tick_params(axis="both", which="major", length=15, width=1.6, labelsize=11)

    plt.tight_layout()

    pdf_path = os.path.join(fig_dir, f"{fig_name}.pdf")
    eps_path = os.path.join(fig_dir, f"{fig_name}.eps")

    if fig is None:
        plt.savefig(pdf_path, format="pdf")
        plt.savefig(eps_path, format="eps")
        plt.show()
    else:
        fig.savefig(pdf_path, format="pdf")
        fig.savefig(eps_path, format="eps")
        plt.show()


# Illustrator-friendly text in PDFs
mpl.rcParams["pdf.fonttype"] = 42
mpl.rcParams["ps.fonttype"]  = 42
mpl.rcParams["svg.fonttype"] = "none"

def style_axes(ax, spine_lw=1.8, tick_len=8, tick_w=1.6, labelsize=11):
    for s in ["bottom","left"]:
        ax.spines[s].set_linewidth(spine_lw)
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.tick_params(axis="both", which="major", length=tick_len, width=tick_w, labelsize=labelsize)
df = nfl_plot_data.copy()
df = df.dropna(subset=["personid", "days_after_lkn", "log_nfl", "mrs_90"]).copy()
df["mrs_90"] = df["mrs_90"].astype(int)  # 0–6
day_order = sorted(df["days_after_lkn"].unique())

mrs_cmap = plt.cm.viridis
norm = mpl.colors.Normalize(vmin=0, vmax=6)

# ============================================================
# 1) Faceted scatter: log(NfL) vs mRS (by day)
# ============================================================
g = sns.FacetGrid(
    df,
    col="days_after_lkn",
    col_order=day_order,
    col_wrap=4,
    height=3.0,
    aspect=1.1,
    sharey=True
)

def scatter_by_mrs(data, **kwargs):
    ax = plt.gca()
    ax.scatter(
        data["mrs_90"] + np.random.uniform(-0.15, 0.15, size=len(data)),
        data["log_nfl"],
        c=data["mrs_90"],
        cmap=mrs_cmap,
        norm=norm,
        s=28,
        alpha=0.8,
        edgecolor="black",
        linewidth=0.4
    )

g.map_dataframe(scatter_by_mrs)

for ax, day in zip(g.axes.flatten(), day_order):
    ax.set_title(f"Day {day}")
    ax.set_xlabel("mRS-90")
    ax.set_ylabel("log(NfL)")
    style_axes(ax)

# colorbar for FacetGrid
cax = g.fig.add_axes([0.92, 0.25, 0.02, 0.5])
cb = mpl.colorbar.ColorbarBase(cax, cmap=mrs_cmap, norm=norm, orientation="vertical")
cb.set_label("mRS-90")

g.fig.suptitle("log(NfL) vs Ordinal mRS, Stratified by Day", y=1.02)
g.fig.tight_layout(rect=[0, 0, 0.9, 1])

finalize_and_save("lognfl_by_mrs_scatter_by_day", fig=g.fig)

# ============================================================
# 2) Day-wise Spearman/Kendall
# ============================================================
rows = []
for day, d in df.groupby("days_after_lkn"):
    if d["mrs_90"].nunique() < 2 or d["log_nfl"].notna().sum() < 3:
        rows.append([day, len(d), np.nan, np.nan, np.nan, np.nan])
        continue
    rho, p_rho = spearmanr(d["log_nfl"], d["mrs_90"], nan_policy="omit")
    tau, p_tau = kendalltau(d["log_nfl"], d["mrs_90"], nan_policy="omit")
    rows.append([day, len(d), rho, p_rho, tau, p_tau])

corr_by_day = pd.DataFrame(
    rows,
    columns=["day","n","spearman_rho","spearman_p","kendall_tau","kendall_p"]
).sort_values("day")
print(corr_by_day)

# Heatmap vector
vec = corr_by_day.set_index("day")["spearman_rho"].to_frame().T

plt.figure(figsize=(10, 1.8))
ax = sns.heatmap(
    vec, cmap="coolwarm", center=0, vmin=-1, vmax=1,
    cbar_kws={"label": "Spearman ρ (logNfL vs mRS)"}
)
ax.set_yticks([0.5])
ax.set_yticklabels(["ρ"])
ax.set_xlabel("Days after LKN")
ax.set_title("Day-wise Association Between log(NfL) and Ordinal mRS")
finalize_and_save("spearman_lognfl_mrs_by_day_heatmap")

# Line plot
plt.figure(figsize=(7,4))
plt.plot(corr_by_day["day"], corr_by_day["spearman_rho"], marker="o")
plt.axhline(0, linestyle="--", linewidth=1)
plt.xlabel("Days after LKN")
plt.ylabel("Spearman ρ")
plt.title("Day-wise Spearman Association (logNfL vs mRS)")
style_axes(plt.gca())
finalize_and_save("spearman_lognfl_mrs_by_day_line")

# ============================================================
# 3) Subject-level features
# ============================================================
def subject_features(df, auc_window=(0,5)):
    out = []

    def fit_slope(dsub):
        if len(dsub) >= 2 and dsub["days_after_lkn"].nunique() >= 2:
            X = np.vstack([np.ones(len(dsub)), dsub["days_after_lkn"].values]).T
            y = dsub["log_nfl"].values
            beta, *_ = lstsq(X, y, rcond=None)
            return beta[1]
        return np.nan

    for pid, d in df.groupby("personid"):
        d = d.sort_values("days_after_lkn")
        mrs = int(d["mrs_90"].iloc[0])

        baseline_log_nfl = d["log_nfl"].iloc[0]
        baseline_day = d["days_after_lkn"].iloc[0]
        peak_log_nfl = d["log_nfl"].max()

        slope_0_1 = fit_slope(d[(d["days_after_lkn"] >= 0) & (d["days_after_lkn"] <= 1)])
        slope_1_3 = fit_slope(d[(d["days_after_lkn"] >= 1) & (d["days_after_lkn"] <= 3)])
        slope_0_3 = fit_slope(d[(d["days_after_lkn"] >= 0) & (d["days_after_lkn"] <= 3)])

        d_a = d[(d["days_after_lkn"] >= auc_window[0]) & (d["days_after_lkn"] <= auc_window[1])]
        if len(d_a) >= 2 and d_a["days_after_lkn"].nunique() >= 2:
            auc_0_5 = np.trapz(d_a["log_nfl"].values, d_a["days_after_lkn"].values)
        else:
            auc_0_5 = np.nan

        out.append([pid, mrs, baseline_day, baseline_log_nfl, peak_log_nfl,
                    slope_0_1, slope_1_3, slope_0_3, auc_0_5])

    return pd.DataFrame(out, columns=[
        "personid","mrs_90","baseline_day","baseline_log_nfl","peak_log_nfl",
        "slope_0_1","slope_1_3","slope_0_3","auc_0_5"
    ])

feat = subject_features(df)
print(feat.head())

feature_list = ["baseline_log_nfl","peak_log_nfl","slope_0_1","slope_1_3","slope_0_3","auc_0_5"]

def plot_feature_vs_mrs(feat, feature, title=None):
    d = feat.dropna(subset=["mrs_90", feature]).copy()

    plt.figure(figsize=(8,5))
    x = d["mrs_90"]
    y = d[feature]

    plt.scatter(
        x + np.random.uniform(-0.15, 0.15, size=len(x)),
        y,
        c=x, cmap=mrs_cmap, norm=norm,
        s=45, alpha=0.85,
        edgecolor="black", linewidth=0.5
    )

    plt.xlabel("90-day mRS")
    plt.ylabel(feature)
    plt.title(title if title else f"{feature} vs 90-day mRS")

    cb = plt.colorbar()
    cb.set_label("mRS-90")

    finalize_and_save(f"subject_level_{feature}_vs_mrs")

def plot_median_iqr(feat, feature):
    d = feat.dropna(subset=["mrs_90", feature]).copy()

    s = d.groupby("mrs_90")[feature].agg(
        median="median",
        q25=lambda x: x.quantile(0.25),
        q75=lambda x: x.quantile(0.75),
        n="count"
    ).reset_index()

    plt.figure(figsize=(7,4))

    plt.scatter(
        d["mrs_90"] + np.random.uniform(-0.12, 0.12, size=len(d)),
        d[feature],
        c=d["mrs_90"], cmap=mrs_cmap, norm=norm,
        s=35, alpha=0.6,
        edgecolor="black", linewidth=0.4
    )

    plt.plot(s["mrs_90"], s["median"], color="black", linewidth=2, marker="o")
    plt.fill_between(s["mrs_90"], s["q25"], s["q75"], color="black", alpha=0.15)

    plt.xlabel("90-day mRS")
    plt.ylabel(feature)
    plt.title(f"{feature}: median (IQR) with individual subjects")

    cb = plt.colorbar()
    cb.set_label("mRS-90")

    finalize_and_save(f"subject_level_{feature}_median_iqr")

for f in feature_list:
    plot_feature_vs_mrs(feat, f, title=f"{f.replace('_',' ').title()} vs 90-day mRS")
    plot_median_iqr(feat, f)

# ============================================================
# 4) Correlations for subject-level features
# ============================================================
tests = []
for feature in feature_list:
    sub = feat.dropna(subset=[feature, "mrs_90"])
    rho, p = spearmanr(sub[feature], sub["mrs_90"])
    tau, p_tau = kendalltau(sub[feature], sub["mrs_90"])
    tests.append([feature, len(sub), rho, p, tau, p_tau])

assoc = pd.DataFrame(tests, columns=["feature","n","spearman_rho","spearman_p","kendall_tau","kendall_p"])
assoc["spearman_p_fdr"] = multipletests(assoc["spearman_p"], method="fdr_bh")[1]
print(assoc.sort_values("spearman_p_fdr"))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

import pandas as pd
from scipy.stats import mannwhitneyu
import statsmodels.api as sm
import statsmodels.formula.api as smf
from scipy.stats import spearmanr, pearsonr
sns.set_style('ticks')
mpl.rcParams["pdf.fonttype"] = 42   # TrueType fonts (editable)
mpl.rcParams["ps.fonttype"] = 42
mpl.rcParams["svg.fonttype"] = "none"
# ----------------------------------------
# 1. Create outcome group
# ----------------------------------------
nfl_plot_data['mrs_30_bin'] = nfl_plot_data['mrs_30'].apply(
    lambda x: "Good (0–3)" if x <= 3 else "Poor (4–6)"
)

# consistent palette
palette = {
    "Good (0–3)": sns.color_palette("tab10")[0],
    "Poor (4–6)": sns.color_palette("tab10")[1],
}

# ----------------------------------------
# 2. Jitter + outcome-based x separation
# ----------------------------------------
jitter_width = 0.10   # small jitter for readability
group_offset = 0.15   # separation between groups

rng = np.random.default_rng(42)
nfl_plot_data_current = nfl_plot_data.copy()
nfl_plot_data_current = nfl_plot_data_current.dropna(subset=['mrs_30'])
# jitter
nfl_plot_data_current['x_jitter'] = rng.uniform(
    -jitter_width, jitter_width, size=len(nfl_plot_data_current)
)

# group offset
group_map = {
    "Good (0–3)": -group_offset,
    "Poor (4–6)": +group_offset
}

nfl_plot_data_current['x_plot'] = nfl_plot_data_current.apply(
    lambda row: row['days_after_lkn'] +
                row['x_jitter'] +
                group_map[row['mrs_30_bin']],
    axis=1
)

# ----------------------------------------
# 3. Summary Counts (subjects per day per outcome)
# ----------------------------------------
summary_counts = (
    nfl_plot_data_current.groupby(['days_after_lkn', 'mrs_30_bin'])['personid']
    .nunique()
    .reset_index()
    .pivot(index='days_after_lkn', columns='mrs_30_bin', values='personid')
)

print("\nNumber of subjects per day by outcome group:")
print(summary_counts.fillna(0))

# ----------------------------------------
# 4. Compute group mean ± SD (use true day positions)
# ----------------------------------------
summary_stats = nfl_plot_data_current.groupby(['days_after_lkn', 'mrs_30_bin']).agg(
    mean=('log_nfl', 'mean'),
    sd=('log_nfl',  'std')
).reset_index()

# add x-position offset for group separation (WITHOUT jitter)
summary_stats['x_mean'] = summary_stats.apply(
    lambda row: row['days_after_lkn'] + group_map[row['mrs_30_bin']],
    axis=1
)

# ----------------------------------------
# 5. Plot
# ----------------------------------------
plt.figure(figsize=(11, 6))

# A. Jittered raw points
sns.scatterplot(
    data=nfl_plot_data_current,
    x='x_plot', y='log_nfl',
    hue='mrs_30_bin',
    palette=palette,
    s=55,
    alpha=0.85,
    linewidth=0.5,
    edgecolor='black'
)

# B. Individual subject lines (perfect alignment)
for (pid, group), dfsub in nfl_plot_data_current.groupby(['personid', 'mrs_30_bin']):
    plt.plot(
        dfsub['x_plot'],
        dfsub['log_nfl'],
        color=palette[group],
        alpha=0.25,
        linewidth=1
    )

# C. Mean ± SD (plotted at true timepoints + group offset)
for group, dfsub in summary_stats.groupby('mrs_30_bin'):
    col = palette[group]

    # mean line
    plt.plot(
        dfsub['x_mean'], dfsub['mean'],
        color=col,
        linewidth=3
    )

    # SD ribbon
    plt.fill_between(
        dfsub['x_mean'],
        dfsub['mean'] - dfsub['sd'],
        dfsub['mean'] + dfsub['sd'],
        color=col,
        alpha=0.18
    )

# ----------------------------------------
# 6. Formatting
# ----------------------------------------
plt.title("Longitudinal NfL Values by Functional Outcome (mRS-30)", fontsize=16)
plt.xlabel("Days After Stroke", fontsize=12)
plt.ylabel("log(NfL)", fontsize=12)
ax = plt.gca()

# Thicken visible spines
for spine in ["bottom", "left"]:
    ax.spines[spine].set_linewidth(1.8)

# Optionally hide top/right spines (clean, journal style)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.tick_params(
    axis="both",
    which="major",
    length=15,      # tick length
    width=1.6,     # tick thickness
    labelsize=11
)

# Legend formatting
plt.legend(title="Outcome", frameon=True)

plt.tight_layout()
fig_dir = '/Users/ganshengtan/Library/CloudStorage/Box-Box/Washu/projects/taVNS_cytokine/stroke-taVNS/nfl_study/figs/'

pdf_path = os.path.join(fig_dir, f"mrs_30_nfl.pdf")
eps_path = os.path.join(fig_dir, f"mrs_30_nfl.eps")
plt.savefig(pdf_path, format="pdf")
plt.savefig(eps_path, format="eps")
plt.show()




print("gee, dependent 1 is good")
gee = smf.gee(
    "mrs_30_bin ~ log_nfl",
    groups=nfl_plot_data_current["personid"],
    data=nfl_plot_data_current,
    family=sm.families.Binomial()
).fit()

print(gee.summary())

# nfl_plot_data["mrs_30_bin_num"] = nfl_plot_data["mrs_30"].apply(lambda x: 0 if x <= 3 else 1)
nfl_plot_data["mrs_30_bin_num"] = np.where(
    nfl_plot_data["mrs_30"].isna(),
    np.nan,
    (nfl_plot_data["mrs_30"] > 3).astype(int)
)
logit = smf.logit("mrs_30_bin_num ~ log_nfl", nfl_plot_data.dropna(subset=["log_nfl", "mrs_30"])).fit()
print(logit.summary())

# daywise_results = pd.DataFrame(daywise_results,
#     columns=["day", "n_good", "n_poor", "p_value"])

# print(daywise_results)


clean = nfl_plot_data[['log_nfl','mrs_30', 'mrs_30_bin', 'mrs_30_bin_num', 'days_after_lkn']].dropna(subset=["log_nfl", "mrs_30"])

rho, p_rho = spearmanr(clean['log_nfl'], clean['mrs_30_bin_num'])
r, p_r  = pearsonr(clean['log_nfl'], clean['mrs_30_bin_num'])

print("Spearman:", rho, p_rho)

rho_all, p_s_all = spearmanr(clean["log_nfl"], clean["mrs_30"])
r_all, p_p_all = pearsonr(clean["log_nfl"], clean["mrs_30"])

print("\n=== Overall Correlations ===")
print(f"Spearman rho = {rho_all:.3f}, p = {p_s_all:.4f}")
print(f"Pearson r = {r_all:.3f}, p = {p_p_all:.4f}")

daywise_corrs = []

for day, df_day in clean.groupby("days_after_lkn"):
    if df_day["mrs_30"].nunique() > 1:
        rho, p_s = spearmanr(df_day["log_nfl"], df_day["mrs_30"])
        r, p_p = pearsonr(df_day["log_nfl"], df_day["mrs_30"])
    else:  # cannot compute correlation if all mRS same
        rho, p_s, r, p_p = np.nan, np.nan, np.nan, np.nan

    daywise_corrs.append([day, len(df_day), rho, p_s, r, p_p])

daywise_corrs = pd.DataFrame(
    daywise_corrs,
    columns=["day", "n", "Spearman_rho", "Spearman_p", "Pearson_r", "Pearson_p"]
)

print("\n=== Day-wise Correlations (log_nfl vs mRS 30) ===")
print(daywise_corrs)

from scipy.stats import mannwhitneyu, ttest_ind, pointbiserialr
import numpy as np

results = []

for day, df_day in nfl_plot_data.groupby("days_after_lkn"):

    df_day = df_day.dropna(subset=['log_nfl', 'mrs_30_bin_num'])
    g = df_day[df_day['mrs_30_bin_num'] == 0]['log_nfl']
    p = df_day[df_day['mrs_30_bin_num'] == 1]['log_nfl']

    # Skip day if any group has <2 subjects
    if len(g) < 2 or len(p) < 2:
        results.append([day, len(g), len(p),
                        np.nan, np.nan, np.nan, np.nan, np.nan, np.nan])
        continue

    # ------------------------------------------------------
    # A. Mann–Whitney U
    # ------------------------------------------------------
    U, p_mwu = mannwhitneyu(g, p, alternative="two-sided")

    # Rank-biserial correlation
    n1, n2 = len(g), len(p)
    r_rb = 1 - (2 * U) / (n1 * n2)

    # ------------------------------------------------------
    # B. t-test
    # ------------------------------------------------------
    t_stat, p_t = ttest_ind(g, p, equal_var=False)

    # Cohen’s d (Hedges' g optional)
    mean_diff = g.mean() - p.mean()
    pooled_sd = np.sqrt(((n1 - 1)*g.var() + (n2 - 1)*p.var()) / (n1 + n2 - 2))
    d = mean_diff / pooled_sd

    # ------------------------------------------------------
    # C. Point-biserial correlation
    # ------------------------------------------------------
    labels = df_day['mrs_30_bin_num']     # 0 good, 1 poor
    values = df_day['log_nfl']
    r_pb, p_pb = pointbiserialr(labels, values)

    # ------------------------------------------------------
    # Store
    # ------------------------------------------------------
    results.append([
        day, n1, n2,
        U, p_mwu,
        t_stat, p_t,
        d, r_rb, r_pb
    ])

# Final dataframe
results_df = pd.DataFrame(
    results,
    columns=[
        'day','n_good','n_poor',
        'U','p_mwu',
        't_stat','p_t',
        'cohen_d','rank_biserial_r','point_biserial_r'
    ]
)

print("\n=== Day-wise Effect Size Summary ===")
print(results_df)
```

```python
# treat it as longitudinal variable
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.model_selection import KFold
from lifelines.utils import concordance_index
import statsmodels.api as sm

# Assume: nfl_plot_data has columns:
# ["personid", "days_after_lkn", "log_nfl", "mrs_90"]

df = nfl_plot_data.copy()

# 1) Define common time grid
time_grid = np.arange(0, 6, 1.0)  # days 0–5, adjust as needed

# 2) Build subject × time matrix (with interpolation)
subjects = df["personid"].unique()
curves = []
mrs_list = []

for pid in subjects:
    df_sub = df[df["personid"] == pid].sort_values("days_after_lkn")
    if df_sub["mrs_90"].nunique() != 1:
        # sanity check: one mRS per subject
        continue
    mrs_val = df_sub["mrs_90"].iloc[0]

    # interpolate log_nfl over grid
    x = df_sub["days_after_lkn"].values
    y = df_sub["log_nfl"].values

    # require at least 2 timepoints to interpolate
    if len(x) < 2:
        continue

    y_grid = np.interp(time_grid, x, y)
    curves.append(y_grid)
    mrs_list.append(mrs_val)

curves = np.vstack(curves)           # shape: (n_subjects, n_timepoints)
mrs_arr = np.array(mrs_list)        # shape: (n_subjects,)

# 3) FPCA via PCA on centered curves
pca = PCA(n_components=3)            # adjust #components
scores = pca.fit_transform(curves)   # FPCA scores (subject × k)

from sklearn.preprocessing import StandardScaler

X = scores
y = mrs_arr

# Standardize predictors
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Add intercept for statsmodels
X_sm = sm.add_constant(X_scaled)

# Cross-validated concordance index (how well we rank mRS)
kf = KFold(n_splits=5, shuffle=True, random_state=42)
c_indices = []

for train_idx, test_idx in kf.split(X_sm):
    X_tr, X_te = X_sm[train_idx], X_sm[test_idx]
    y_tr, y_te = y[train_idx], y[test_idx]

    model = sm.OLS(y_tr, X_tr).fit()
    y_pred = model.predict(X_te)

    c_idx = concordance_index(y_te, y_pred)  
    # note: lower mRS = better; use -y_pred so higher "risk" = worse outcome
    c_indices.append(c_idx)

print("Cross-validated c-index (mean ± SD):",
      np.mean(c_indices), np.std(c_indices))

def permutation_cindex(X, y, n_perm=1000, n_splits=5):
    rng = np.random.default_rng(42)
    kf = KFold(n_splits=n_splits, shuffle=True, random_state=42)

    def cv_cindex(X, y):
        cis = []
        for tr, te in kf.split(X):
            model = sm.OLS(y[tr], X[tr]).fit()
            preds = model.predict(X[te])
            cis.append(concordance_index(y[te], preds))
        return np.mean(cis)

    real_c = cv_cindex(X, y)
    perm_c = []

    for _ in range(n_perm):
        y_perm = rng.permutation(y)
        perm_c.append(cv_cindex(X, y_perm))

    perm_c = np.array(perm_c)
    p_val = np.mean(perm_c >= real_c)
    return real_c, p_val, perm_c

real_c, p_val, perm_dist = permutation_cindex(X_sm, y)
print(f"Real CV c-index = {real_c:.3f}, permutation p = {p_val:.4f}")

import matplotlib.pyplot as plt

plt.hist(perm_dist, bins=30, alpha=0.7, label="Permutation null")
plt.axvline(real_c, color="red", linewidth=2, label=f"Real c-index = {real_c:.3f}")
plt.axvline(0.5, color="black", linestyle="--", label="Chance (c=0.5)")

plt.xlabel("Cross-validated c-index")
plt.ylabel("Frequency")
plt.title("Permutation Test for FPCA→mRS Prediction")
plt.legend()
plt.show()

plt.scatter(scores[:,0], scores[:,1], 
            c=mrs_arr, cmap="viridis", s=80, edgecolor="k")

plt.colorbar(label="mRS 90")
plt.xlabel("FPCA1 score")
plt.ylabel("FPCA2 score")
plt.title("FPCA Components of NfL Trajectories\nColored by 90-day mRS")
plt.show()


components = pca.components_     # shape: (n_components, n_timepoints)
explained = pca.explained_variance_ratio_
plt.figure(figsize=(10,5))

for i in range(3):   # assuming 3 components
    plt.plot(time_grid, components[i], marker="o", label=f"PC{i+1} ({explained[i]*100:.1f}%)")

plt.axhline(0, color="black", linewidth=0.5)
plt.xlabel("Days after LKN")
plt.ylabel("Loading Weight")
plt.title("PCA Component Loadings (Eigen-Trajectories)")
plt.legend()
plt.show()
```

```python
# get nihss information from the nihss_df
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from scipy.stats import mannwhitneyu
import statsmodels.api as sm
import statsmodels.formula.api as smf
from scipy.stats import spearmanr, pearsonr
imputed_nfl_df = pd.read_csv(
    '/Users/ganshengtan/Library/CloudStorage/Box-Box/Washu/projects/taVNS_cytokine/stroke-taVNS/nfl_study/organized_data/nihss_nfl_cyto_mrs_wbc_mrs_imputedct.csv',
    index_col=0
)
sns.set(
    style="ticks",
    context="paper",
    font_scale=1.1
)
nfl_plot_data = (
    imputed_nfl_df[imputed_nfl_df['days_after_lkn'].isin([0,1,3,5])]
    .dropna(subset=['nfl'])
    .drop_duplicates(subset=['personid','days_after_lkn','nfl'])
    .copy()
)
nfl_plot_data['smoking_binary'] = nfl_plot_data['smoking'].map({0: 0, 1: 1, 2: 1})

nfl_plot_data['log_nfl'] = np.log(nfl_plot_data['nfl'])
nfl_plot_data.columns

# use nfl to predict mrs 30, mrs_90, nihss over time

data_filename = '/Users/ganshengtan/Library/CloudStorage/Box-Box/Washu/projects/taVNS_cytokine/stroke-taVNS/Copy of NUVISTA 13 Nov 2024.xlsx'
nihss_df = pd.read_excel(data_filename, sheet_name='nihss - analysis file')
nihss_30_df = (
    nihss_df[nihss_df["timepoint_nihss"] == 30]
    .groupby("personid")["nihss"]
    .first()
    .reset_index()
    .rename(columns={"nihss": "nihss_30"})
)
binary_thres  = np.median(nihss_30_df['nihss_30'])
# Step 2 — merge into nfl_plot_data
nfl_plot_data = nfl_plot_data.merge(nihss_30_df, on="personid", how="left")




# ----------------------------------------
# 1. Create outcome group
# ----------------------------------------
nfl_plot_data['nihss_30_bin'] = nfl_plot_data['nihss_30'].apply(
    lambda x: f"Good (<={binary_thres})" if x <= binary_thres else f"Poor (>{binary_thres})"
)

# consistent palette
palette = {
    f"Good (<={binary_thres})": sns.color_palette("tab10")[0],
    f"Poor (>{binary_thres})": sns.color_palette("tab10")[1],
}

# ----------------------------------------
# 2. Jitter + outcome-based x separation
# ----------------------------------------
jitter_width = 0.10   # small jitter for readability
group_offset = 0.15   # separation between groups

rng = np.random.default_rng(42)
nfl_plot_data_current = nfl_plot_data.copy()
nfl_plot_data_current = nfl_plot_data_current.dropna(subset=['nihss_30'])
# jitter
nfl_plot_data_current['x_jitter'] = rng.uniform(
    -jitter_width, jitter_width, size=len(nfl_plot_data_current)
)

# group offset
group_map = {
    f"Good (<={binary_thres})": -group_offset,
    f"Poor (>{binary_thres})": +group_offset
}

nfl_plot_data_current['x_plot'] = nfl_plot_data_current.apply(
    lambda row: row['days_after_lkn'] +
                row['x_jitter'] +
                group_map[row['nihss_30_bin']],
    axis=1
)

# ----------------------------------------
# 3. Summary Counts (subjects per day per outcome)
# ----------------------------------------
summary_counts = (
    nfl_plot_data_current.groupby(['days_after_lkn', 'nihss_30_bin'])['personid']
    .nunique()
    .reset_index()
    .pivot(index='days_after_lkn', columns='nihss_30_bin', values='personid')
)

print("\nNumber of subjects per day by outcome group:")
print(summary_counts.fillna(0))

# ----------------------------------------
# 4. Compute group mean ± SD (use true day positions)
# ----------------------------------------
summary_stats = nfl_plot_data_current.groupby(['days_after_lkn', 'nihss_30_bin']).agg(
    mean=('log_nfl', 'mean'),
    sd=('log_nfl',  'std')
).reset_index()

# add x-position offset for group separation (WITHOUT jitter)
summary_stats['x_mean'] = summary_stats.apply(
    lambda row: row['days_after_lkn'] + group_map[row['nihss_30_bin']],
    axis=1
)

# ----------------------------------------
# 5. Plot
# ----------------------------------------
plt.figure(figsize=(11, 6))

# A. Jittered raw points
sns.scatterplot(
    data=nfl_plot_data_current,
    x='x_plot', y='log_nfl',
    hue='nihss_30_bin',
    palette=palette,
    s=55,
    alpha=0.85,
    linewidth=0.5,
    edgecolor='black'
)

# B. Individual subject lines (perfect alignment)
for (pid, group), dfsub in nfl_plot_data_current.groupby(['personid', 'nihss_30_bin']):
    plt.plot(
        dfsub['x_plot'],
        dfsub['log_nfl'],
        color=palette[group],
        alpha=0.25,
        linewidth=1
    )

# C. Mean ± SD (plotted at true timepoints + group offset)
for group, dfsub in summary_stats.groupby('nihss_30_bin'):
    col = palette[group]

    # mean line
    plt.plot(
        dfsub['x_mean'], dfsub['mean'],
        color=col,
        linewidth=3
    )

    # SD ribbon
    plt.fill_between(
        dfsub['x_mean'],
        dfsub['mean'] - dfsub['sd'],
        dfsub['mean'] + dfsub['sd'],
        color=col,
        alpha=0.18
    )

# ----------------------------------------
# 6. Formatting
# ----------------------------------------
plt.title("Longitudinal NfL Values by Functional Outcome (nihss-30)", fontsize=16)
plt.xlabel("Days After Stroke", fontsize=12)
plt.ylabel("log(NfL)", fontsize=12)

# Legend formatting
plt.legend(title="Outcome", frameon=True)

plt.tight_layout()
finalize_and_save(f"nihss_30_log_nfl_acorss_day")
plt.show()



print("gee, dependent 1 is good")
gee = smf.gee(
    "nihss_30_bin ~ log_nfl",
    groups=nfl_plot_data_current["personid"],
    data=nfl_plot_data_current,
    family=sm.families.Binomial()
).fit()

print(gee.summary())
# greater nihss_30 = bin = 1, poor 
nfl_plot_data["nihss_30_bin_num"] = np.where(
    nfl_plot_data["nihss_30"].isna(),
    np.nan,
    (nfl_plot_data["nihss_30"] > binary_thres).astype(int)
)
logit = smf.logit("nihss_30_bin_num ~ log_nfl", nfl_plot_data.dropna(subset=["log_nfl", "nihss_30"])).fit()
print(logit.summary())

# daywise_results = pd.DataFrame(daywise_results,
#     columns=["day", "n_good", "n_poor", "p_value"])

# print(daywise_results)


clean = nfl_plot_data[['log_nfl','nihss_30', 'nihss_30_bin', 'nihss_30_bin_num', 'days_after_lkn']].dropna(subset=["log_nfl", "nihss_30"])

rho, p_rho = spearmanr(clean['log_nfl'], clean['nihss_30_bin_num'])
r, p_r  = pearsonr(clean['log_nfl'], clean['nihss_30_bin_num'])

print("Spearman:", rho, p_rho)

rho_all, p_s_all = spearmanr(clean["log_nfl"], clean["nihss_30"])
r_all, p_p_all = pearsonr(clean["log_nfl"], clean["nihss_30"])

print("\n=== Overall Correlations ===")
print(f"Spearman rho = {rho_all:.3f}, p = {p_s_all:.4f}")
print(f"Pearson r = {r_all:.3f}, p = {p_p_all:.4f}")

daywise_corrs = []

for day, df_day in clean.groupby("days_after_lkn"):
    if df_day["nihss_30"].nunique() > 1:
        rho, p_s = spearmanr(df_day["log_nfl"], df_day["nihss_30"])
        r, p_p = pearsonr(df_day["log_nfl"], df_day["nihss_30"])
    else:  # cannot compute correlation if all mRS same
        rho, p_s, r, p_p = np.nan, np.nan, np.nan, np.nan

    daywise_corrs.append([day, len(df_day), rho, p_s, r, p_p])

daywise_corrs = pd.DataFrame(
    daywise_corrs,
    columns=["day", "n", "Spearman_rho", "Spearman_p", "Pearson_r", "Pearson_p"]
)

print("\n=== Day-wise Correlations (log_nfl vs nihss30) ===")
print(daywise_corrs)

from scipy.stats import mannwhitneyu, ttest_ind, pointbiserialr
import numpy as np

results = []

for day, df_day in nfl_plot_data.groupby("days_after_lkn"):

    df_day = df_day.dropna(subset=['log_nfl', 'nihss_30_bin_num'])
    g = df_day[df_day['nihss_30_bin_num'] == 0]['log_nfl']
    p = df_day[df_day['nihss_30_bin_num'] == 1]['log_nfl']

    # Skip day if any group has <2 subjects
    if len(g) < 2 or len(p) < 2:
        results.append([day, len(g), len(p),
                        np.nan, np.nan, np.nan, np.nan, np.nan, np.nan])
        continue

    # ------------------------------------------------------
    # A. Mann–Whitney U
    # ------------------------------------------------------
    U, p_mwu = mannwhitneyu(g, p, alternative="two-sided")

    # Rank-biserial correlation
    n1, n2 = len(g), len(p)
    r_rb = 1 - (2 * U) / (n1 * n2)

    # ------------------------------------------------------
    # B. t-test
    # ------------------------------------------------------
    t_stat, p_t = ttest_ind(g, p, equal_var=False)

    # Cohen’s d (Hedges' g optional)
    mean_diff = g.mean() - p.mean()
    pooled_sd = np.sqrt(((n1 - 1)*g.var() + (n2 - 1)*p.var()) / (n1 + n2 - 2))
    d = mean_diff / pooled_sd

    # ------------------------------------------------------
    # C. Point-biserial correlation
    # ------------------------------------------------------
    labels = df_day['nihss_30_bin_num']     # 0 good, 1 poor
    values = df_day['log_nfl']
    r_pb, p_pb = pointbiserialr(labels, values)

    # ------------------------------------------------------
    # Store
    # ------------------------------------------------------
    results.append([
        day, n1, n2,
        U, p_mwu,
        t_stat, p_t,
        d, r_rb, r_pb
    ])

# Final dataframe
results_df = pd.DataFrame(
    results,
    columns=[
        'day','n_good','n_poor',
        'U','p_mwu',
        't_stat','p_t',
        'cohen_d','rank_biserial_r','point_biserial_r'
    ]
)

print("\n=== Day-wise Effect Size Summary ===")
print(results_df)

# pearson across days
Y_LIM = (0, 7)
days = [0, 1, 3, 5]
plt.rcParams.update({
    "pdf.fonttype": 42,   # TrueType fonts (editable in Illustrator)
    "ps.fonttype": 42,
    "svg.fonttype": "none"
})
# keep only rows with NIHSS-30 and log_nfl
scatter_df = nfl_plot_data_current.dropna(
    subset=["log_nfl", "nihss_30"]
).copy()

fig, axes = plt.subplots(
    1, len(days),
    figsize=(14, 4),
    sharey=True
)

for ax, day in zip(axes, days):
    sub = scatter_df[scatter_df["days_after_lkn"] == day]

    if len(sub) < 3:
        ax.set_title(f"Day {day}\n(n<3)")
        ax.axis("off")
        continue

    # scatter
    sns.scatterplot(
        data=sub,
        x="nihss_30",
        y="log_nfl",
        ax=ax,
        s=60,
        alpha=0.8,
        edgecolor="black",
        linewidth=0.5
    )

    # correlations
    rho, p_s = spearmanr(sub["nihss_30"], sub["log_nfl"])
    r, p_p   = pearsonr(sub["nihss_30"], sub["log_nfl"])

    ax.text(
            0.05, 0.95,
            f"Spearman ρ = {rho:.2f}\nPearson r = {r:.2f}",
            transform=ax.transAxes,
            ha="left", va="top",
            fontsize=10,
            bbox=dict(facecolor="white", alpha=0.85, edgecolor="none")
        )
    
    ax.set_title(f"Day {day}")
    ax.set_xlabel("NIHSS (Day 30)")
    ax.set_ylim(Y_LIM)
    ax.tick_params(
        axis="both",
        which="major",
        direction="out",
        length=8,        # longer ticks
        width=1.6,       # thicker ticks
        labelsize=11
        )
    
    ax.tick_params(
        axis="both",
        which="minor",
        direction="out",
        length=4,
        width=1.2
    )
    
    # Only left & bottom spines
    for spine in ["top", "right"]:
        ax.spines[spine].set_visible(False)
    
    for spine in ["left", "bottom"]:
        ax.spines[spine].set_linewidth(1.8)
axes[0].set_ylabel("log(NfL)", fontsize=12)
sns.despine()
plt.tight_layout()
finalize_and_save("nihss30_vs_lognfl_scatter_by_day")
plt.show()
```

```python
nfl_plot_data.columns


# We can predict those / or correlation
# ("dispos_good", dispos_good_mapping, dispos_good_levels, dispos_good_palette),
# ("readmission", readmission_mapping, readmission_levels, readmission_palette),

# get nihss information from the nihss_df
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import mannwhitneyu
import statsmodels.api as sm
import statsmodels.formula.api as smf
from scipy.stats import spearmanr, pearsonr
sns.set(style="whitegrid")

# ----------------------------------------
# 1. Create outcome group
# ----------------------------------------
# dispo_good==0 means poor or died
nfl_plot_data["dispos_good"] = nfl_plot_data["dispos_good"].astype(int)
nfl_plot_data["dispos_good_bin"] = nfl_plot_data["dispos_good"].map({
    1: "Good",
    0: "Poor"
})
# consistent palette
palette = {
    "Good": sns.color_palette("tab10")[0],
    "Poor": sns.color_palette("tab10")[1],
}

# ----------------------------------------
# 2. Jitter + outcome-based x separation
# ----------------------------------------
jitter_width = 0.10   # small jitter for readability
group_offset = 0.15   # separation between groups

rng = np.random.default_rng(42)
nfl_plot_data_current = nfl_plot_data.copy()
nfl_plot_data_current = nfl_plot_data_current.dropna(subset=['dispos_good_bin'])
# jitter
nfl_plot_data_current['x_jitter'] = rng.uniform(
    -jitter_width, jitter_width, size=len(nfl_plot_data_current)
)

# group offset
group_map = {
    "Good": -group_offset,
    "Poor": +group_offset
}

nfl_plot_data_current['x_plot'] = nfl_plot_data_current.apply(
    lambda row: row['days_after_lkn'] +
                row['x_jitter'] +
                group_map[row['dispos_good_bin']],
    axis=1
)

# ----------------------------------------
# 3. Summary Counts (subjects per day per outcome)
# ----------------------------------------
summary_counts = (
    nfl_plot_data_current.groupby(['days_after_lkn', 'dispos_good_bin'])['personid']
    .nunique()
    .reset_index()
    .pivot(index='days_after_lkn', columns='dispos_good_bin', values='personid')
)

print("\nNumber of subjects per day by outcome group:")
print(summary_counts.fillna(0))

# ----------------------------------------
# 4. Compute group mean ± SD (use true day positions)
# ----------------------------------------
summary_stats = nfl_plot_data_current.groupby(['days_after_lkn', 'dispos_good_bin']).agg(
    mean=('log_nfl', 'mean'),
    sd=('log_nfl',  'std')
).reset_index()

# add x-position offset for group separation (WITHOUT jitter)
summary_stats['x_mean'] = summary_stats.apply(
    lambda row: row['days_after_lkn'] + group_map[row['dispos_good_bin']],
    axis=1
)

# ----------------------------------------
# 5. Plot
# ----------------------------------------
plt.figure(figsize=(11, 6))

# A. Jittered raw points
sns.scatterplot(
    data=nfl_plot_data_current,
    x='x_plot', y='log_nfl',
    hue='dispos_good_bin',
    palette=palette,
    s=55,
    alpha=0.85,
    linewidth=0.5,
    edgecolor='black'
)

# B. Individual subject lines (perfect alignment)
for (pid, group), dfsub in nfl_plot_data_current.groupby(['personid', 'dispos_good_bin']):
    plt.plot(
        dfsub['x_plot'],
        dfsub['log_nfl'],
        color=palette[group],
        alpha=0.25,
        linewidth=1
    )

# C. Mean ± SD (plotted at true timepoints + group offset)
for group, dfsub in summary_stats.groupby('dispos_good_bin'):
    col = palette[group]

    # mean line
    plt.plot(
        dfsub['x_mean'], dfsub['mean'],
        color=col,
        linewidth=3
    )

    # SD ribbon
    plt.fill_between(
        dfsub['x_mean'],
        dfsub['mean'] - dfsub['sd'],
        dfsub['mean'] + dfsub['sd'],
        color=col,
        alpha=0.18
    )

# ----------------------------------------
# 6. Formatting
# ----------------------------------------
plt.title("Longitudinal NfL Values by Functional Outcome (mRS-90)", fontsize=16)
plt.xlabel("Days After Stroke", fontsize=12)
plt.ylabel("log(NfL)", fontsize=12)

# Legend formatting
plt.legend(title="Outcome", frameon=True)

plt.tight_layout()
plt.show()



print("gee, dependent 1 is good")
gee = smf.gee(
    "dispos_good ~ log_nfl",
    groups=nfl_plot_data_current["personid"],
    data=nfl_plot_data_current,
    family=sm.families.Binomial()
).fit()

print(gee.summary())

logit = smf.logit("dispos_good ~ log_nfl", nfl_plot_data.dropna(subset=["log_nfl", "dispos_good"])).fit()
print(logit.summary())

daywise_results = pd.DataFrame(daywise_results,
    columns=["day", "n_good", "n_poor", "p_value"])

print(daywise_results)


clean = nfl_plot_data[['log_nfl','dispos_good', 'days_after_lkn']].dropna(subset=["log_nfl", "dispos_good"])

rho, p_rho = spearmanr(clean['log_nfl'], clean['dispos_good'])
r, p_r  = pearsonr(clean['log_nfl'], clean['dispos_good'])

print("Spearman:", rho, p_rho)

rho_all, p_s_all = spearmanr(clean["log_nfl"], clean["dispos_good"])
r_all, p_p_all = pearsonr(clean["log_nfl"], clean["dispos_good"])

print("\n=== Overall Correlations ===")
print(f"Spearman rho = {rho_all:.3f}, p = {p_s_all:.4f}")
print(f"Pearson r = {r_all:.3f}, p = {p_p_all:.4f}")

daywise_corrs = []

for day, df_day in clean.groupby("days_after_lkn"):
    if df_day["dispos_good"].nunique() > 1:
        rho, p_s = spearmanr(df_day["log_nfl"], df_day["dispos_good"])
        r, p_p = pearsonr(df_day["log_nfl"], df_day["dispos_good"])
    else:  # cannot compute correlation if all mRS same
        rho, p_s, r, p_p = np.nan, np.nan, np.nan, np.nan

    daywise_corrs.append([day, len(df_day), rho, p_s, r, p_p])

daywise_corrs = pd.DataFrame(
    daywise_corrs,
    columns=["day", "n", "Spearman_rho", "Spearman_p", "Pearson_r", "Pearson_p"]
)

print("\n=== Day-wise Correlations (log_nfl vs mRS 90) ===")
print(daywise_corrs)

from scipy.stats import mannwhitneyu, ttest_ind, pointbiserialr
import numpy as np

results = []

for day, df_day in nfl_plot_data.groupby("days_after_lkn"):

    df_day = df_day.dropna(subset=['log_nfl', 'dispos_good'])
    g = df_day[df_day['dispos_good'] == 1]['log_nfl']
    p = df_day[df_day['dispos_good'] == 0]['log_nfl']

    # Skip day if any group has <2 subjects
    if len(g) < 2 or len(p) < 2:
        results.append([day, len(g), len(p),
                        np.nan, np.nan, np.nan, np.nan, np.nan, np.nan])
        continue

    # ------------------------------------------------------
    # A. Mann–Whitney U
    # ------------------------------------------------------
    U, p_mwu = mannwhitneyu(g, p, alternative="two-sided")

    # Rank-biserial correlation
    n1, n2 = len(g), len(p)
    r_rb = 1 - (2 * U) / (n1 * n2)

    # ------------------------------------------------------
    # B. t-test
    # ------------------------------------------------------
    t_stat, p_t = ttest_ind(g, p, equal_var=False)

    # Cohen’s d (Hedges' g optional)
    mean_diff = g.mean() - p.mean()
    pooled_sd = np.sqrt(((n1 - 1)*g.var() + (n2 - 1)*p.var()) / (n1 + n2 - 2))
    d = mean_diff / pooled_sd

    # ------------------------------------------------------
    # C. Point-biserial correlation
    # ------------------------------------------------------
    labels = df_day['dispos_good']     # 0 good, 1 poor
    values = df_day['log_nfl']
    r_pb, p_pb = pointbiserialr(labels, values)

    # ------------------------------------------------------
    # Store
    # ------------------------------------------------------
    results.append([
        day, n1, n2,
        U, p_mwu,
        t_stat, p_t,
        d, r_rb, r_pb
    ])

# Final dataframe
results_df = pd.DataFrame(
    results,
    columns=[
        'day','n_good','n_poor',
        'U','p_mwu',
        't_stat','p_t',
        'cohen_d','rank_biserial_r','point_biserial_r'
    ]
)

print("\n=== Day-wise Effect Size Summary ===")
print(results_df)
```
